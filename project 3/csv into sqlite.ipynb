```python
# import pandas
import pandas as pd

# import sql alchemy
from sqlalchemy import create_engine

# import pymysql
import pymysql
pymysql.install_as_MySQLdb()
```

```python
# import and establish Base for which classes will be constructed
from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
```

```python
# import modules to declare columns and column data types
from sqlalchemy import Column, Integer, String, Float
```

```python
# create 311 ticket class

class Tickets(Base):
    __tablename__ = '311_tickets'
    id = Column(Integer, primary_key=True)
    source = Column(String(255))
    department = Column(String(255))
    work_group = Column(String(255))
    request_type = Column(String(255))
    category = Column(String(255))
    type = Column(String(255))
    detail = Column(String(255))
    creation_date = Column(Integer)
    creation_time = Column(String(255))
    creation_month = Column(Integer)
    creation_year = Column(Integer)
    status = Column(String(255))
    exceeded_est_timeframe = Column(String(255))
    closed_date = Column(Integer)
    closed_month = Column(Integer)
    closed_year = Column(Integer)
    days_to_close = Column(Integer)
    street_address = Column(String(255))
    address_with_geocode = Column(String(255))
    zip_code = Column(String(255))
    neighborhood = Column(String(255))
    county = Column(String(255))
    council_district = Column(String(255))
    police_district = Column(String(255))
    parcel_id_no = Column(Integer)
    ycoordinate = Column(Integer)
    xcoordinate = Column(Integer)
    case_url = Column(String(255))
    days_open = Column(Integer)
```

```python
from sklearn import tree
import pandas as pd
import os
```

```python
df = tickets_311_df
```

```python
target = df["DAYS TO CLOSE"]
target_names = ["negative", "positive"]
```

```python
data = df.drop(["DAYS TO CLOSE","CASE ID","SOURCE","DEPARTMENT","WORK GROUP","REQUEST TYPE","CATEGORY","TYPE","DETAIL","CREATION DATE","CREATION TIME","CREATION MONTH","CREATION YEAR","STATUS","EXCEEDED EST TIMEFRAME"], axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
# WARNING! BOILERPLATE CODE HERE! 
# Use this to visualize the tree
import graphviz 
dot_data = tree.export_graphviz(
    clf, out_file=None, 
    feature_names=feature_names,  
    class_names=target_names,  
    filled=True, rounded=True,  
    special_characters=True)  

graph = graphviz.Source(dot_data)  
graph 
```

```python
 from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
